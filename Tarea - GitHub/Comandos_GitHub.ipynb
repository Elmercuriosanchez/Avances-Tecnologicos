1. **¿Qué es GitHub?**

GitHub es una plataforma web que funciona como un servicio de alojamiento de repositorios de código en la nube. Su principal propósito es facilitar la colaboración y la gestión de proyectos de software a través de Internet. A menudo se le describe como una "red social para desarrolladores".

Es crucial entender que GitHub no es el sistema de control de versiones en sí mismo, sino una plataforma construida sobre él. El software fundamental que utiliza es Git.

**Git vs. GitHub: La diferencia esencial**

 Git es el Sistema de Control de Versiones Distribuido (un software que se instala localmente). Permite rastrear los cambios en tu código a lo largo del tiempo, crear diferentes líneas de trabajo (ramas) y revertir a versiones anteriores.

 GitHub es la plataforma de hosting. Proporciona la infraestructura centralizada en la nube para almacenar los repositorios de Git y añade herramientas sociales y de gestión, como la revisión de código a través de Pull Requests y el seguimiento de errores (Issues).

2. **Comandos Básicos de Git (para interactuar con GitHub)**

 Los comandos de GitHub son los comandos de Git que ejecutas en tu terminal para gestionar el código de forma local y sincronizarlo con el repositorio remoto alojado en GitHub.

**A. Primeros pasos y configuración**

 Para empezar a trabajar con Git y GitHub, primero debes configurar tu identidad y crear o descargar el proyecto:

 - **git config --global user.name "Tu Nombre":** Establece tu nombre de usuario para que aparezca en tus contribuciones.

 - **git config --global user.email "tu@email.com":** Establece tu correo electrónico asociado a tus commits.

 - **git init:** Inicializa un nuevo repositorio Git en la carpeta actual. Es el punto de partida de cualquier proyecto local.

 - **git clone [url]:** Descarga una copia completa de un repositorio existente de GitHub (o cualquier servidor Git) a tu máquina.

**B. El ciclo de vida de los cambios**

El flujo de trabajo básico implica modificar archivos, prepararlos y guardarlos:

Revisar el estado:

 - **git status:** Muestra qué archivos han sido modificados, cuáles están listos para guardar, y cuáles son nuevos.

Preparar los cambios:

 - **git add [archivo]** o **git add** .: Mueve los archivos modificados o nuevos al "área de preparación" (staging area), marcándolos para el próximo commit.

 **Guardar (Commit):**

 - **git commit -m "Mensaje descriptivo":** Registra los cambios preparados de forma permanente en el historial de tu repositorio local. El mensaje es fundamental para documentar lo que hiciste.

**C. Colaboración y sincronización (Push/Pull)**

 Estos comandos son el puente entre tu trabajo local y el repositorio remoto en GitHub:

 - **git remote add origin [url]:** Vincula tu repositorio local con la URL del repositorio en GitHub (normalmente se hace solo una vez).

 - **git push -u origin main:** Sube (envía) tus commits locales a la rama remota en GitHub. Es la forma de compartir tu trabajo.

 - **git pull origin main:** Descarga y fusiona los últimos cambios del repositorio remoto de GitHub a tu rama local. Es crucial para mantener tu código actualizado con el trabajo de tus colaboradores.

**D. Gestión de Ramas (Branches)**

 Las ramas permiten crear entornos de desarrollo aislados para nuevas funcionalidades sin interrumpir la línea principal del código:

 - **git branch [nombre-rama]:** Crea una nueva rama.

 - **git checkout [nombre-rama]:** Cambia tu entorno de trabajo a la rama especificada.

 - **git merge [nombre-rama]:** Fusiona el historial de cambios de la rama especificada en la rama en la que te encuentras actualmente.

 - **git log:** Muestra el historial completo de commits de tu rama.
